```python
import model_example  

(x_train, y_train), (x_test, y_test) = model_example.get_dataset()
model = model_example.get_model()
```

```python
import tensorflow as tf

def scheduler(epoch, lr):
    return 1e-6 * 10**(epoch / 30)

lrs_callback = tf.keras.callbacks.LearningRateScheduler(scheduler)

history = model.fit(x=x_train, y=y_train, epochs=200, validation_data=(x_test, y_test), callbacks=[lrs_callback])
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(2, 1, 1)
ax.plot(history.history["lr"], history.history['val_loss'])
ax.set_xscale('log')
plt.show()
```
